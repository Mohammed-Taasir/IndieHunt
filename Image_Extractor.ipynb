```python
df = pd.read_csv('Test.csv')
image_dict = dict()
img_save_flag = 1
os.mkdir('temp_dir')
```

```python
count = 0
for i in range(len(df)):
  for k in range(0,20):
    col_name = "imageUrls_" + str(k)
    url = df[col_name][i]
    if url != '0' and url != '' and url != ' ':
      try:
        resp=urllib.request.urlopen(url)
        img_arr=np.array(bytearray(resp.read()),dtype='uint8')
        img=cv2.imdecode(img_arr,cv2.IMREAD_COLOR)
        img_new=cv2.resize(img,(224,224),interpolation= cv2.INTER_AREA)
        image_dict[count] = [df['place_id'][i], img_new, df['SubcategoryName'][i]]
        if img_save_flag:
          img_name = '_'.join(df['SubcategoryName'][i].split(' ')) + "_" + str(count)
          cv2.imwrite("/content/temp_dir/" + img_name + ".jpg", img_new)
        count += 1
      except:
        print("could not reach image")
```

```python
import pickle as pkl

fobj=open('train_dict1.pkl','wb')
pkl.dump(image_dict,fobj)
fobj.close()
files.download('train_dict1.pkl')
```
