**Review Tags**

```python
cols_to_merge = ['reviewsTags_0_title','reviewsTags_1_title', 'reviewsTags_2_title', 'reviewsTags_3_title', 'reviewsTags_4_title', 'reviewsTags_5_title', 'reviewsTags_6_title', 'reviewsTags_7_title', 'reviewsTags_8_title', 'reviewsTags_9_title']
df['reviewTagsList'] = df.apply(lambda x: ' '.join(set(col for col in x.iloc[21:29] if col != '0')), axis=1)

# # drop the original columns
# df.drop(cols_to_merge, axis=1, inplace=True)

# print the merged column
print(df['reviewTagsList'])
```

```python
def Union(lst1, lst2):
    final_list = lst1 + lst2
    return final_list

for i in range(df.shape[0]):
  list_a = df.iloc[i]['reviewTagsList'].split(' ')
  list_b = df.iloc[i]['desc_list'].split(' ')
  set_ = set(Union(list_a,list_b))
  new_s = {x for x in set_ if x != ''}
  result = ' '.join(new_s)
  df.at[i, 'final_feature_set'] = result
```

```python
tf = pd.DataFrame(df, columns = ['place_id','final_feature_set'])
```

```python
def jaccard_similarity(query_set,tag_set):
        union_set=query_set.union(tag_set)
        intersct=query_set.intersection(tag_set)
        jac_sim=len(intersct)/len(union_set)
        return jac_sim

```

```python
input_query = 'environment dating party picnic feel'
query_set = set(input_query.split())
for i in range(tf.shape[0]):
  word_set = set(tf.iloc[i]['final_feature_set'].split())
  tf.at[i,'jacc_simi_'] = jaccard_similarity(query_set,word_set)


tf = tf.sort_values('jacc_simi_', ascending=False)
top_100_ranked_places = []
top_100_ranked_places = tf[:100]
```
