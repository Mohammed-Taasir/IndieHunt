```python
import json
import csv

def flatten_json(nested_json):
    """
    Flatten a nested JSON object
    """
    out = {}

    def flatten(x, name=''):
        if type(x) is dict:
            for a in x:
                flatten(x[a], name + a + '_')
        elif type(x) is list:
            i = 0
            for a in x:
                flatten(a, name + str(i) + '_')
                i += 1
        else:
            out[name[:-1]] = x

    flatten(nested_json)
    return out

"""
Convert a JSON file to CSV file
"""
json_file = "delhi-tourist-places.json"
with open(json_file, 'r') as f:
    data = json.load(f)
  
csv_file = "output.csv"

flattened_data = []
for item in data:
    flattened_item = flatten_json(item)
    flattened_data.append(flattened_item)

with open(csv_file, 'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=flattened_data[0].keys())
    writer.writeheader()
    for item in flattened_data:
        writer.writerow(item)

# Example usage
```

```python
import csv
import json

def flatten_json(nested_json, exclude_keys=[]):
    """
    Flatten a nested JSON object with different properties into a flat dictionary.
    """
    out = {}
    
    def flatten(data, prefix='', exclude_keys=[]):
        if type(data) is dict:
            for key in data:
                if key not in exclude_keys:
                    flatten(data[key], prefix + key + '_', exclude_keys)
        elif type(data) is list:
            for i, item in enumerate(data):
                flatten(item, prefix + str(i) + '_', exclude_keys)
        else:
            out[prefix[:-1]] = data
    
    flatten(nested_json, exclude_keys=exclude_keys)
    return out
def convert_json_to_csv(json_file, csv_file, exclude_keys=[]):
    """
    Convert a JSON file to a CSV file.
    """
    with open(json_file, 'r', encoding='utf-8') as f:
        data = json.load(f)
    
    # Flatten the nested JSON data
    flattened_data = [flatten_json(item, exclude_keys=exclude_keys) for item in data]
    
    # Get the list of all keys
    keys = set().union(*(d.keys() for d in flattened_data))
    
    # Write the data to a CSV file
    with open(csv_file, 'w', newline='', encoding='utf-8') as f:
        writer = csv.DictWriter(f, fieldnames=keys)
        writer.writeheader()
        writer.writerows(flattened_data)


```

```python
import pandas as pd
import glob

# Step 1: Get a list of all CSV files in the directory
csv_files = glob.glob("*.csv")

# Step 2: Create an empty list to store the data from all CSV files
merged_data = []

# Step 3: Loop through the list of CSV files and read them into dataframes
for file in csv_files:
    df = pd.read_csv(file)

    # Step 4: Merge the dataframes
    merged_data.append(df)

# Step 5: Concatenate the merged dataframes into one dataframe
merged_df = pd.concat(merged_data, sort=False)

# Step 6: Write the merged data to a new CSV file
merged_df.to_csv("merged.csv", index=False)
```
