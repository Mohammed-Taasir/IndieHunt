## Extract the images and store as jpg, and build a dict to store data about the images

```python
image_dict = dict()
os.mkdir('temp_dir')
```

```python
count = 0
for i in range(len(df)):
  score = df['final_score'][i]
  if score > 1:
    for k in range(0,20):
      col_name = "imageUrls_" + str(k)
      url = df[col_name][i]
      if url != '0' and url != '' and url != ' ':
        try:
          resp=urllib.request.urlopen(url)
          img_arr=np.array(bytearray(resp.read()),dtype='uint8')
          img=cv2.imdecode(img_arr,cv2.IMREAD_COLOR)
          img_new=cv2.resize(img,(224,224),interpolation= cv2.INTER_AREA)
          img_name = '_'.join(df['SubcategoryName'][i].split(' ')) + "_1_" + str(count)
          cv2.imwrite("/content/temp_dir/" + img_name + ".jpg", img_new)
          image_dict[img_name] = [df['place_id'][i], url, df['final_score'][i]]
          count += 1
        except:
          print("could not reach image : row = ",i)
```

## Download the extracted images and the Dictionary

```python
import pickle as pkl
from zipfile import ZipFile

fobj=open('train_dict1.pkl','wb')
pkl.dump(image_dict,fobj)
fobj.close()
files.download('train_dict1.pkl')

files_list=os.listdir('/content/temp_dir')
dir_path='/content/temp_dir'
zf = ZipFile("images_set1.zip", "w")
for dirname, subdirs, fles in os.walk(dir_path):
    zf.write(dirname)
    for filename in fles:
        zf.write(os.path.join(dirname, filename))
zf.close()
files.download('/content/images_set1.zip')
```
